```python
import plotly.express as px
import pandas as pd
import numpy as numpy
import plotly.graph_objects as go

df=pd.read_csv('tatamotorsstockprices.csv')

# Atribuí o gráfico a uma variável para facilitar a inclusão de novos métodos na visualização
fig = px.line(
    df,
    x='Date',
    y='Adj Close',
    labels={'Date':'Data', 'Adj Close':'Fechamento Ajustado'},
    title='Preço das Ações - Tata Motors'
)

# Adicionando efeitos
fig.update_xaxes(
    # Mostrando o efeito sanfona no gráfico
    rangeslider_visible=True,
    # Adicionando filtro temporal
    rangeselector=dict(
        # Adicionando botões para realizar o filtro temporal
        buttons=list([
            # count - Quantidade de passos que serão dados
            # label - O texto que vai aparecer no botão
            # step - Aquilo que será "passado". Pode ser ano, dia, mês, semana, etc. A quantidade de passos está em step
            # stepmode - Configura se o filtro temporal vai filtrar para trás ou até o momento presente 
            dict(count=1, label='1m', step='month', stepmode='backward'),
            dict(count=6, label='6m', step='month', stepmode='backward'),
            dict(count=1, label='YTD', step='year', stepmode='todate'),
            dict(count=1, label='1y', step='year', stepmode='backward')
        ])
    )
    )

# Mostrando o gráfico
fig.show()
```

```python
fig = go.Figure(data=go.Candlestick(
    x=df['Date'],
    open=df['Open'],
    high=df['High'],
    low=df['Low'],
    close=df['Close']    
    )
)

fig.update_layout(
    title='Tata Motors Stock Prices',
    yaxis_title='Stock Prices',
    xaxis_title='Data'
)

fig.show()
```
