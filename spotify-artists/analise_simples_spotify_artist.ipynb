```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as mp
```

```python
df = pd.read_csv('./spotify-artists/Artists.csv', sep=',')
print(df.head(5))
```

```python
mp.pie(
        df['Gender'].value_counts(),
        labels=("Masculino", "Feminino", "Não definido", "Outros"),
        wedgeprops={"linewidth": 1, "edgecolor": "white"},
        shadow=True,
        autopct='%.2f%%'
)
mp.title('Artistas por Gênero (US)')
mp.show()
```

```python
df2 = df[df['Age'] > 0]
df2.sort_values(by='Age', ascending=True)
```

### Argumentos de Marker
https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers

```python
mp.scatter(df2['Age'], df2['Popularity'], marker='>', color='#ff0050', )
```

```python
x = df2['Followers']/10000
y = df2['Popularity']

# label vai ser utilizado no método legend()
# linestyle funciona somente para gráficos com linha
# color aceita nomes de cores e hexadecimal
mp.plot(x, y, label='Seguidores/Popularidade', linestyle = 'dashed', color='#ff0000', marker='v', linewidth='0.5')
mp.ylabel('Popularidade')
mp.xlabel('Seguidores')
mp.title('Popularidade por Seguidor')
mp.xticks(np.arange(0, 12000, 1300))
mp.legend()
mp.show()
```

```python
x = [1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 70, 80, 90, 100]
y = [22, 57, 30, 122, 220, 80, 43, 22, 10, 95, 95, 12, 125, 30, 27, 15, 110]

# linefmt aceita como argumento nomes de cores e estilos predefinidos de C1 a C9;
# linefmt também aceita concatenado à cor, o linestyle como mostrado abaixo ":" tornando a linha pontilhada
# ATENÇÃO: Se a cor for escrita, o comando acima não funciona. Quer testar? Coloca 'Red:' em linefmt e vê se funciona
# Bottom buga os valores de y, pois funciona como linha base.
#   Ex: Bottom=0 x=1 y=22 > Padrão
#       Bottom=5 x=1 y=27* > Bottom editado
# A linha de Bottom é a linha base dos valores de y, eles serão sempre y+bottom no valor final
# No gráfico de haste (stem) o x é chamado de locs e y de heads (e y só aceita array como argumento)

mp.stem(x, y, linefmt='C3:', markerfmt='8', label='Precipitação', bottom=0, orientation='vertical')
mp.ylabel('Precipitação (mm)')
mp.xlabel('Dias')
# mp.xticks(x) - Aplica os dados de x ao eixo x
# mp.yticks(y) - Aplica os dados de y ao eixo y
mp.grid(True)
mp.legend()
mp.show()
```
