```python
import pandas as pd
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('japan_birth.csv')

fig = px.line(
    df,
    x='year',
    y=['population_male', 'population_female'],
    labels={'year':'Ano', 'value':'População total', 'variable':'Sexo'},
    title='Análise Populacional - Japão'
)

fig.update_xaxes(rangeslider_visible=False)
fig.show()
```

```python
fig = px.bar(
    df,
    x='year',
    y=['population_male', 'population_female'],
    labels={'value':'População Total', 'year':'Ano', 'variable':'Gênero'},
    title='Análise Demográfica da População'
)

fig.show()
```
