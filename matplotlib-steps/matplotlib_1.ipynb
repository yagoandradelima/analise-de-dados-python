```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Gráfico de linha
plt.plot([1, 2, 3, 4])
plt.ylabel('Números')
plt.show()

# Sintaxe mínima:
# plt.plot(eixo_y, eixo_x)
# Quando eixo_x não é declarado o python automaticamente coloca um array de 0 até o comprimento de dados do eixo_y
# Para esse caso o array ficou: eixo_x = [0, 1, 2, 3]
```

```python
# Novo exemplo de plt.plot declarando os dois eixos
plt.plot([1, 2, 3, 4], [3, 7, 9, 12])
plt.ylabel('Números')
plt.xlabel('Eixo x')
plt.show()
```

```python
# 'ro' é um comando de estilo conjunto que é baseado no MATLAB
# 'r' = red (vermelho) - Cor da linha
# 'o' = circle (círculo) - Tipo de linha

plt.plot([1, 4, 6, 0], [12, 4, 5, 2], 'ro')
plt.ylabel('Eixo y')
plt.xlabel('Eixo x')
plt.show()
```

```python
# As propriedades do gráfico são bem intuitivas, a tradução delas literalmente significa sua propriedade
plt.plot(
         [1, 2, 3, 4], 
         [5, 8, 9, 12], 
         linestyle='--', 
         color='#ff0000', 
         marker='p', 
         markersize=9,
         markeredgecolor='green',
         markeredgewidth=2
)

plt.show()
```

```python
# Dá para plotar o mesmo gráfico várias vezes e alterar seu estilo utilizando fmt
t = np.arange(0., 5., 0.2)

# plt.plot(eixo_x, eixo_y, estilo: vermelho tracejado, eixo_x2, eixo_y2, estilo: azul círculos, eixo_x3, eixo_y3, estilo: verde triângulos)
plt.plot(t, t, 'r--', t, t**2, 'bo', t, t**3, 'g^')
plt.show()
```
