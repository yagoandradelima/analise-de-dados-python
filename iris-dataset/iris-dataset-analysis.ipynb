```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('iris-dataset.csv')
df.head(15)
```

```python
plt.style.use('seaborn')
plt.scatter(
            df['sepal_length'], # eixo_x
            df['sepal_width'], # eixo_y
            c=df['petal_width'], # Filtro de cores (sÃ³ aceita nÃºmeros ou arrays de nÃºmeros) 
            cmap='Blues', # Cores do filtro de cores
            s=df['petal_length']*20, # Tamanho dos marcadores
            edgecolor='black', # Cor da borda dos marcadores
            linewidth=1 # Espessura da borda dos marcadores
)

plt.colorbar().set_label('Espessura da PÃ©tala') # Criando e colocando a Legenda da barra de cores
plt.xlabel('Comprimento da SÃ©pala') # Legenda do eixo_x
plt.ylabel('Espessura da SÃ©pala') # Legenda do eixo_y
# plt.grid()
```
