```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("rice_ countries.csv")

print(df)

df = pd.melt(
    df, # Dataframe que quero realizar a operação
    id_vars=['Country'], # Coluna que será meu index para a transformação (ficará fixa)
    value_vars=['1990', '2000', '2010', '2020'], # Colunas que serão transformadas em linhas
    var_name='Year', # Nome da coluna que será criada com o index das colunas como dados
    value_name='Crops' # Nome da coluna de valores
)
print('-'*15)
print(df.head(5))
```

```python
plt.figure(figsize=(17,4))
plt.plot(df['Year'], df['Crops'], label=df['Country'])
plt.legend()
```
